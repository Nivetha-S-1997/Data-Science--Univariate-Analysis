```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
dataset
```

```python
Quan=[]
Qual=[]
for columnName in dataset.columns:
    if dataset[columnName].dtype=="O":
        #print("Qual")
        Qual.append(columnName)
    else:
        #print("Quan")
        Quan.append(columnName)
```

```python
Quan
```

```python
Qual
```

```python
Univariate=pd.DataFrame(index=["Mean","Median","Mode"],columns=Quan)
for columnName in Quan:
    Univariate[columnName]["Mean"]=dataset[columnName].mean()
    Univariate[columnName]["Median"]=dataset[columnName].median()
    Univariate[columnName]["Mode"]=dataset[columnName].mode()[0]
```

```python
Univariate
```

```python
def CentralTendency(dataset):
    Univariate=pd.DataFrame(index=["Mean","Median","Mode"],columns=Quan)
    for columnName in Quan:
        Univariate[columnName]["Mean"]=dataset[columnName].mean()
        Univariate[columnName]["Median"]=dataset[columnName].median()
        Univariate[columnName]["Mode"]=dataset[columnName].mode()[0]
    return Univariate
```
