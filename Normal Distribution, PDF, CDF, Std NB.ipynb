# Probability Density Function

```python
def pdf(dataset,startrange,endrange): #Defining a function to determine the probabiity density function
    sns.distplot(dataset,kde_kws={'color':'Green'},color='Blue') #Using seaborn library, normal distribution curve and probability density function curves are drawn
    pyplot.axvline(startrange,color='Red') #Using pyplot function, vertical lines in the specified range of density functions are drawn
    pyplot.axvline(endrange,color='Red')
    
    Mean=dataset.mean() #Mean of the given column in the dataset is calculated
    Std_dev=dataset.std() #Standard Deviation of the given column in the dataset is calculated
    print("Mean: ",Mean, "Standard deviation: ", Std_dev) #Printing the calculated values of mean and standard deviation
    
    dist=norm(Mean,Std_dev) #Using the calculated mean and standard deviation values, distribution is calculated using the norm function
    
    values=[value for value in range(startrange,endrange)] #Using 1-liner for loop, each value in the specified range is get as a list of values
    probability=[dist.pdf(value) for value in values] #Probability density function are calculated for each values in the given range
    prob_func=sum(probability) #Summing up all the probabilities to get the final probability in the specified range
    print("The area between {} & {} is {}".format(startrange,endrange,prob_func)) #Printing the final probability density value
    return prob_func
```

```python
pdf(dataset["ssc_p"],70,80)
```

```python
def pdf(dataset,startrange,endrange):
    sns.distplot(dataset,kde_kws={'color':'Green'},color='Blue')
    pyplot.axvline(startrange,color='Red')
    pyplot.axvline(endrange,color='Red')
    
    Mean=dataset.mean()
    Std_dev=dataset.std()
    print("Mean: ",Mean, "Standard deviation: ", Std_dev)
    
    dist=norm(Mean,Std_dev)
    
    values=[]
    for value in range(startrange,endrange):
        values.append(value)
        
    probability=[]
    for value in values:
        pdf=dist.pdf(value)
        probability.append(pdf)
    
    prob_func=sum(probability)
    print("The area between {} & {} is {}".format(startrange,endrange,prob_func))
    return prob_func
```

```python
pdf(dataset["ssc_p"],70,80)
```

# Standard Normal Distribution

```python
def stdNB(dataset):
    Mean=dataset.mean() #Mean of the given column in the dataset is calculated
    Std_dev=dataset.std() #Standard Deviation of the given column in the dataset is calculated
    print("Mean: ",Mean, "Standard Deviation: ",Std_dev) #Printing the values of Mean and Standard Deviation
    
    values=[i for i in dataset] #Getting all the values in the input column as a list using one liner for loop
    z_score=[((x-Mean)/Std_dev) for x in values] #Calculating z-score for each value in the specified column
    sns.distplot(z_score,kde_kws={"color":"Red"}) #Plotting the standard normal distribution graph
    #return z_score
```

```python
stdNB(dataset["ssc_p"])
```
